# What is the probability of a game selling 1 million copies if it is made by  X360 and from the Action genre, vs. the probability if it is made by WiiU and from the Strategy genre?

```python
def prob(platform, genre):
    return len(df[(df['Genre'] == genre) & (df['Platform'] == platform)]) / len(df[df['Global_Sales'] >= 1.0])
```

```python
print(prob('X360', 'Action'))
print(prob('WiiU', 'Strategy'))
```

```python
probs = {'Popular vs Unpopular': ['Action Games in X360', 'Strategy Games in WiiU'], 'Probabilities': [0.15569437770302738
, 0.0014416146083613647]}
         
probs_df = pd.DataFrame(probs, columns=['Popular vs Unpopular', 'Probabilities'])

sns.barplot(x='Popular vs Unpopular', y='Probabilities', data=probs_df)

         
         
         
```
